```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = 'cleaned_dataset.csv'
data = pd.read_csv(file_path)

#Scatter Plot
plt.figure(figsize=(8, 6))
sns.scatterplot(data=data, x='bill_length_mm', y='bill_depth_mm', hue='species')
plt.title('Scatter Plot: Bill Length vs. Bill Depth')
plt.xlabel('Bill Length (mm)')
plt.ylabel('Bill Depth (mm)')
plt.legend(title='Species')
plt.show()

#Line Plot
plt.figure(figsize=(8, 6))
sns.lineplot(data=data.sort_values('bill_length_mm'), x='bill_length_mm', y='bill_depth_mm', hue='species')
plt.title('Line Plot: Bill Length vs. Bill Depth')
plt.xlabel('Bill Length (mm)')
plt.ylabel('Bill Depth (mm)')
plt.legend(title='Species')
plt.show()

#Hexbin Plot
plt.figure(figsize=(8, 6))
plt.hexbin(data['bill_length_mm'], data['bill_depth_mm'], gridsize=30, cmap='Blues')
plt.title('Hexbin Plot: Bill Length vs. Bill Depth')
plt.xlabel('Bill Length (mm)')
plt.ylabel('Bill Depth (mm)')
plt.colorbar(label='Count')
plt.show()

#KDE Plot
plt.figure(figsize=(8, 6))
sns.kdeplot(data=data, x='bill_length_mm', y='bill_depth_mm', hue='species', fill=True)
plt.title('KDE Plot: Bill Length vs. Bill Depth')
plt.xlabel('Bill Length (mm)')
plt.ylabel('Bill Depth (mm)')
species_labels = data['species'].unique()  
plt.legend(title='Species', labels=species_labels)
plt.show()

#Boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(data=data, x='species', y='bill_length_mm', hue='species', palette='pastel', legend=False)
plt.title('Boxplot: Bill Length by Species')
plt.xlabel('Species')
plt.ylabel('Bill Length (mm)')
plt.show()


#Violin Plot
plt.figure(figsize=(8, 6))
sns.violinplot(data=data, x='species', y='bill_depth_mm', hue='species', palette='muted', legend=False)
plt.title('Violin Plot: Bill Depth by Species')
plt.xlabel('Species')
plt.ylabel('Bill Depth (mm)')
plt.show()

#Pair Plot
sns.pairplot(data, vars=['bill_length_mm', 'bill_depth_mm'], hue='species', diag_kind='kde')
plt.suptitle('Pair Plot: Bill Length and Depth', y=1.02)
plt.show()
```
